# **Data Preprocessing**

```python
import numpy as np
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt 

df = pd.read_csv("dataset/breast-cancer.csv")
```

## **Dropping ID Column**

```python
df.drop('id', axis=1, inplace=True)
```

## **Exploratory Data Analysis**

```python
# descriptive statistics of data
df.describe()
```

```python
# countplot
plt.figure(figsize = (8,7))
sns.countplot(x="diagnosis", data=df, palette='magma')
```

```python
# heatmap
plt.figure(figsize=(20,18))
numeric_df = df.select_dtypes(include=np.number)
sns.heatmap(numeric_df.corr(), annot=True, linewidths=.5, cmap="Purples")
```

```python
# counts of unique rows in the 'diagnosis' column
df['diagnosis'].value_counts()
```

## **Modifying the Dataset**

```python
# mapping categorical values to numerical values
df['diagnosis'] = df['diagnosis'].map({'B':0, 'M':1})
```

```python
df['diagnosis'].value_counts()
```

```python
numeric_columns = df.select_dtypes(include=['float64', 'int64']).columns
df[numeric_columns] = df[numeric_columns].astype(float)
```

```python
columns = [col for col in df.columns if col != 'diagnosis'] + ['diagnosis']
df = df[columns]

df['diagnosis'] = df['diagnosis'].astype(int)
```

## **Exporting Preprocessed Data**

```python
df.to_csv('dataset/preprocessed_breast-cancer.csv', index=False)
print("Preprocessing completed. Data saved to CSV.")
```
